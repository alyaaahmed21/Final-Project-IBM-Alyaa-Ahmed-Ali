## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
df.isna().sum().sort_values(ascending=True)
```

```python
df['price'] = df['price'].str.replace('$','')
df['price'] = df['price'].str.replace(',','')
```

```python
df['price'] = df['price'].astype(float)
```

```python
df.info()
```

```python
numeric_columns = ['lat','long','number of reviews', 'price', 'minimum nights','availability 365','calculated host listings count','reviews per month','Construction year']
for i in numeric_columns:
    df[i] = df[i].fillna( df[i].mean())
```

```python
df.isna().sum().sort_values(ascending=True)
```

```python
object_columns = ['neighbourhood','neighbourhood group','cancellation_policy','instant_bookable','instant_bookable','instant_bookable','service fee','host_identity_verified','review rate number','host name','last review','NAME']
for i in object_columns:
    df[i] = df[i].fillna( df[i].mode()[0])
```

```python
df.isna().sum().sort_values(ascending=True)
```

```python
df.drop(['house_rules','license'], axis = 1, inplace = True)
```

```python
df.isna().sum().sort_values(ascending=True)
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
df.duplicated().sum()
df.drop_duplicates(inplace = True)
```

```python
## Display the total number of records in the dataframe after removing the duplicates.
df.shape
```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`

```python
## Rename the column.
df = df.rename(columns = {'availability 365':'days_booked'})
df.columns
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.
## i did it for price column above
## for service fee
df['service_fee']= df['service_fee'].str.strip('$').str.replace(',', '')
df['service_fee'] = df['service_fee'].astype(float)
```

## Task 5a: Data Visualization (Any Tool)

* List the count of various room types avaliable with Airnb
* Which room type adheres to more strict cancellation policy
* List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals
* List the top 10 neighborhoods in the increasing order of their price with the help of a horizontal bar graph. Which is the cheapest neighborhood.
* List the neighborhoods which offer short term rentals within 10 days. Illustrate with a bar graph
* List the prices with respect to room type using a bar graph and also state your inferences.
* Create a pie chart that shows  distribution of booked days for each neighborhood group .Which neighborhood has the highest booking percentage.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
import matplotlib.pyplot as plt
```

```python
df['room_type'].value_counts().plot(kind='bar', color='coral')
plt.title('Count of Various Room Types on Airbnb')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
df[df['cancellation_policy'] == 'strict']['room_type'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('room type adheres to more strict cancellation policy')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.bar(df.groupby('neighbourhood_group')['price'].mean().reset_index()['neighbourhood_group'], df.groupby('neighbourhood_group')['price'].mean().reset_index()['price'], color='skyblue')
plt.title('Average Prices For Each Neighbourhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.show()
```

```python
most_expensive_group = df.groupby('neighbourhood_group')['price'].mean().idxmax()
most_expensive_group   
```

```python
plt.barh(df.sort_values(by='price').head(10)['neighbourhood'], df.sort_values(by='price').head(10)['price'], color='lightcoral')
plt.title('Top 10 Neighborhoods')
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df[df['days_booked'] == 10]['neighbourhood'], df[df['days_booked'] == 10]['days_booked'], color='green')
plt.title('Neighborhoods Offering Short-term Rentals within 10 Days on Airbnb')
plt.xlabel('Neighborhood')
plt.ylabel('days_booked')
plt.show()
```

```python
df.groupby('room_type')['price'].mean().sort_values().plot(kind='bar', color='RoyalBlue')
plt.title('Average Prices with Respect to Room Type on Airbnb')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
```

```python
total_booked_days = df.groupby('neighbourhood_group')['days_booked'].sum()
booking_percentage = total_booked_days / total_booked_days.sum() * 100
plt.figure(figsize=(8,8))
plt.pie(booking_percentage, labels=booking_percentage.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'lightpink', 'lightsalmon'])
plt.title('Distribution of Booked Days by Neighborhood Group on Airbnb')
plt.legend(title='Neighborhood Group', bbox_to_anchor=(1, 0.5), loc="center left", borderaxespad=0.)
plt.show()
## the highest_booking_neighborhood from chart is 'Manhattan' 
```

## Task 5c: Data Visualization (Any Tool)

* With the help of box plots illustrate the following
 * Effect of Review Rate number on price
 * Effect of host identity verified on price

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
plt.boxplot([df[df['review_rate_number'] == rate]['price'] for rate in df['review_rate_number'].unique()], labels=df['review_rate_number'].unique())
plt.title('Effect of Review Rate on Price on Airbnb')
plt.xlabel('Review Rate')
plt.ylabel('Price')
```

```python
plt.boxplot([df[df['host_identity_verified'] == status]['price'] for status in df['host_identity_verified'].unique()],
            labels=df['host_identity_verified'].unique())
plt.title('Effect of Host Identity Verification on Price on Airbnb')
plt.xlabel('Host Identity Verified')
plt.ylabel('Price')
```
